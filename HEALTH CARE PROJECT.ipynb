```python
#CHANGING OBJECT FORMAT TO DATE TIME FORMAT

df1["Camp_Start_Date"]=pd.to_datetime(df1["Camp_Start_Date"])
df1["Camp_End_Date"]=pd.to_datetime(df1["Camp_End_Date"])
```

```python
#APPLYING SOME FUNCTIONS
#Camp_Start_Date
df1['Camp_Start_Month']=df1['Camp_Start_Date'].apply(lambda x:x.month)
df1['Camp_Start_Day']=df1['Camp_Start_Date'].apply(lambda x:x.day)
df1['Camp_Start_Quarter']=df1['Camp_Start_Date'].apply(lambda x:x.quarter)

#Camp_End_Date
df1['Camp_End_Month']=df1['Camp_End_Date'].apply(lambda x:x.month)
df1['Camp_End_Day']=df1['Camp_End_Date'].apply(lambda x:x.day)
df1['Camp_End_Quarter']=df1['Camp_End_Date'].apply(lambda x:x.quarter)

#Camp_Duration
df1['Camp_Duration']=(df1['Camp_End_Date']-df1['Camp_Start_Date']).astype('timedelta64[D]')
```
